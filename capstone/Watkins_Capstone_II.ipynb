```python
train_moscowhd = pd.read_csv('~/Desktop/Russian_Housing/train.csv', parse_dates=['timestamp'])
```

```python
test_moscowhd = pd.read_csv('~/Desktop/Russian_Housing/test.csv', parse_dates=['timestamp'])
```

```python
macro_moscowhd = pd.read_csv('~/Desktop/Russian_Housing/macro.csv', parse_dates=['timestamp'])
```

```python
id_test = test_moscowhd.id
```

```python
y_train = train_moscowhd["price_doc"]
x_train = train_moscowhd.drop(["id", "timestamp", "price_doc"], axis=1)
x_test = test_moscowhd.drop(["id", "timestamp"], axis=1)

#can't merge train with test because the kernel run for very long time

for c in x_train.columns:
    if x_train[c].dtype == 'object':
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(x_train[c].values)+list(x_test[c].values)) 
        x_train[c] = lbl.transform(list(x_train[c].values))       
        x_test[c] = lbl.transform(list(x_test[c].values))
        #x_test.drop(c,axis=1,inplace=True)     
```

```python
xgb_params = {
    'eta': 0.04,
    'max_depth': 4,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'silent': 1
}

dtrain = xgb.DMatrix(x_train, y_train)
dtest = xgb.DMatrix(x_test)

```

```python
cv_output = xgb.cv(xgb_params, dtrain, num_boost_round=1000)
# cv_output[['train-rmse-mean', 'test-rmse-mean']].plot()


# cv_output = xgb.cv(xgb_params, dtrain, num_boost_round=1000, early_stopping_rounds=20,
#     verbose_eval=50, show_stdv=False)
cv_output[['cv-train-rmse-mean', 'cv-test-rmse-mean']].plot()
```
